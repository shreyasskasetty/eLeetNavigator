```python
import pandas as pd
```

```python
prob = pd.read_csv('../data/test-problems.csv')
```

```python
prob
```

```python
prob_id = prob['problems'].unique()
```

```python
prob = prob.drop(columns=['description'])
```

```python
prob
```

```python
prob.to_csv('../data/temp_problem.csv')
```

```python
grouped = prob.groupby('problems')
```

```python
data = []
first_grp = 0
columns = ['id' , 'accepted-solutions', 'submissions', 'acceptance-rate', 'topics', 'companies', 'likes', 'no-of-comments', 'difficulty-level']
for group_name, group_data in grouped:
    row = []
    row.append(group_name)
    first_grp = group_data
    first_row = first_grp.iloc[0]
    row.append(first_row['accepted-solutions'])
    row.append(first_row['submissions'])
    row.append(first_row['acceptance-rate'])
    topic = ','.join(first_grp[first_grp['topics'].notna()]['topics'].unique()) 
    row.append(topic)
    companies = ','.join(first_grp[first_grp['companies'].notna()]['companies'].unique())
    row.append(companies)
    row.append(first_row['likes'])
    row.append(first_row['no-of-comments'])
    row.append(first_row['difficulty-level'])
    data.append(row)

filtered_df = pd.DataFrame(data, columns=['id' , 'accepted-solutions', 'submissions', 'acceptance-rate', 'topics', 'companies', 'likes', 'no-of-comments', 'difficulty-level'])
```

```python
filtered_df
```

```python
filtered_df.to_csv('../data/final-problem-data.csv')
```
